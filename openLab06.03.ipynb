```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Dados = pd.read_csv('regiaoOpenLab.csv', ";")
df = pd.DataFrame(Dados)
```

```python
df
```

```python
dfresult = df.dropna(subset=[1,2], axis=1)
dfresult
```

```python
from sklearn.impute import SimpleImputer
import numpy as np

```

```python
median_imputer = SimpleImputer(missing_values=np.nan,strategy='median')
mean_imputer = SimpleImputer(missing_values=np.nan,strategy='mean')
```

```python
dfresult.taxa_aprovacao = median_imputer.fit_transform(dfresult[['taxa_aprovacao']])
dfresult.nota_saeb_lingua_portuguesa = mean_imputer.fit_transform(dfresult[['nota_saeb_lingua_portuguesa']])
dfresult.nota_saeb_matematica = mean_imputer.fit_transform(dfresult[['nota_saeb_matematica']])
dfresult.indicador_rendimento = mean_imputer.fit_transform(dfresult[['indicador_rendimento']])
dfresult.ideb= mean_imputer.fit_transform(dfresult[['ideb']])
```

```python
dfresult
```

```python
#deixando somente os dados necessarios

dfresult.drop(["nota_saeb_media_padronizada", "rede", "ensino","indicador_rendimento","anos_escolares"], axis=1, inplace=True)
```
